# 09 — Push All Adapters to HuggingFace
Run this only after all 6 training notebooks are complete.
Pushes all local adapters to `iwasbinod/` on HuggingFace.

```python
import sys
sys.path.append('/kaggle/working/final_year_proj_finetuning')
from src.utils import hf_login
from unsloth import FastLanguageModel
import os

hf_login(os.environ.get('HF_TOKEN'))

ADAPTERS = [
    ('outputs/adapters/qwen3-8b-nepali-translation-qlora',    'iwasbinod/qwen3-8b-nepali-translation-qlora'),
    ('outputs/adapters/qwen3-8b-nepali-summarization-qlora',  'iwasbinod/qwen3-8b-nepali-summarization-qlora'),
    ('outputs/adapters/qwen3-8b-nepali-qa-qlora',             'iwasbinod/qwen3-8b-nepali-qa-qlora'),
    ('outputs/adapters/llama32-3b-nepali-translation-qlora',  'iwasbinod/llama32-3b-nepali-translation-qlora'),
    ('outputs/adapters/llama32-3b-nepali-summarization-qlora','iwasbinod/llama32-3b-nepali-summarization-qlora'),
    ('outputs/adapters/llama32-3b-nepali-qa-qlora',           'iwasbinod/llama32-3b-nepali-qa-qlora'),
]

for local_dir, repo_id in ADAPTERS:
    if os.path.exists(local_dir):
        print(f'Pushing {local_dir} → {repo_id} ...')
        model, tokenizer = FastLanguageModel.from_pretrained(
            local_dir, max_seq_length=512, load_in_4bit=True
        )
        model.push_to_hub(repo_id)
        tokenizer.push_to_hub(repo_id)
        print(f'Done: https://huggingface.co/{repo_id}')
        del model
        import torch, gc
        gc.collect(); torch.cuda.empty_cache()
    else:
        print(f'[SKIP] {local_dir} not found — run training notebook first')
```
