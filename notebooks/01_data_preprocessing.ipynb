# 01 — Data Preprocessing
Loads, cleans, and formats all 3 datasets. Saves to JSONL for training.

**Datasets:**
- Translation: `Helsinki-NLP/opus-100` (ne-en) → 5,000 train / 500 val + flores200 test
- Summarization: `csebuetnlp/xlsum` (nepali) → 3,000 train / 300 val
- QA: `xquad` (xquad.ne) → ~952 train / ~238 val (80/20 split)

```python
import sys
sys.path.append('/kaggle/working/final_year_proj_finetuning')

from src.data_utils import (
    load_translation_data,
    load_summarization_data,
    load_qa_data,
    save_jsonl
)
from src.utils import set_seed

set_seed(42)
```

```python
# ── TRANSLATION ──────────────────────────────
print('=' * 50)
print('TRANSLATION')
print('=' * 50)
train_tr, val_tr, test_tr = load_translation_data(n_train=5000, n_val=500)

save_jsonl(train_tr, 'data/processed/translation_train.jsonl')
save_jsonl(val_tr,   'data/processed/translation_val.jsonl')
save_jsonl(test_tr,  'data/processed/translation_test.jsonl')

# Save 50 samples for GitHub
save_jsonl(train_tr[:50], 'data/samples/translation_sample.jsonl')

print('\nSample training example:')
print(train_tr[0]['text'][:300])
```

```python
# ── SUMMARIZATION ────────────────────────────
print('=' * 50)
print('SUMMARIZATION')
print('=' * 50)
train_sum, val_sum = load_summarization_data(n_train=3000, n_val=300)

save_jsonl(train_sum, 'data/processed/summarization_train.jsonl')
save_jsonl(val_sum,   'data/processed/summarization_val.jsonl')
save_jsonl(train_sum[:50], 'data/samples/summarization_sample.jsonl')

print('\nSample training example:')
print(train_sum[0]['text'][:300])
```

```python
# ── QA ───────────────────────────────────────
print('=' * 50)
print('QA')
print('=' * 50)
train_qa, val_qa = load_qa_data()

save_jsonl(train_qa, 'data/processed/qa_train.jsonl')
save_jsonl(val_qa,   'data/processed/qa_val.jsonl')
save_jsonl(train_qa[:50], 'data/samples/qa_sample.jsonl')

print('\nSample training example:')
print(train_qa[0]['text'][:300])
```

```python
# ── FINAL SUMMARY ────────────────────────────
print('\n' + '=' * 50)
print('DATASET SUMMARY')
print('=' * 50)
print(f'Translation  → Train: {len(train_tr):,}  Val: {len(val_tr):,}  Test: {len(test_tr):,}')
print(f'Summarization → Train: {len(train_sum):,}  Val: {len(val_sum):,}')
print(f'QA           → Train: {len(train_qa):,}  Val: {len(val_qa):,}')
print('\nAll data saved to data/processed/')
print('Samples saved to data/samples/ (commit these to GitHub)')
```
